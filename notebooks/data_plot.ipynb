```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from tqdm import tqdm

from utils import exclude_id

id2label, _ = exclude_id((8,9,10))

def dataset_plot(df: pd.DataFrame, size=3, alpha=1):
    fig = plt.figure()
    ax = fig.add_subplot()
    ax.set_xlabel('valence')
    ax.set_ylabel('arousal')
    cmap = cm.get_cmap('gist_ncar')
    ax.set_box_aspect(1)
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)
    label2point = {}
    for idx, row in tqdm(df.iterrows()):
        mp = ax.scatter(row['valence'], row['arousal'],
                        alpha=alpha,
                        label=id2label[row['expression']],
                        c=row['expression'],
                        cmap=cmap,
                        vmin=0,
                        vmax=len(id2label),
                        s=size,)
        label2point[id2label[row['expression']]] = mp
    labels, handles = zip(*sorted(label2point.items()))
    legend = ax.legend(handles, labels, loc='upper left', bbox_to_anchor=(1,0.5))
    return fig, legend
```

```python
fig, legend = dataset_plot(val_dataset.df)
fig.savefig(os.path.join('./results/', 'validation_distribution.svg'), bbox_extra_artists=[legend], bbox_inches='tight')
fig.savefig(os.path.join('./results/', 'validation_distribution.png'), bbox_extra_artists=[legend], bbox_inches='tight')
```

```python
from sklearn.cluster import KMeans

num_clusters = 8

kmeans_valaro = KMeans(n_clusters=num_clusters, random_state=0)
```

```python
valaro_df = val_dataset.df[['valence','arousal']]
valaro_df
```

```python
kmeans_valaro.fit(valaro_df)
kmeans_valaro.labels_
```

```python
import numpy as np
import warnings
warnings.simplefilter('ignore')

x_range = np.linspace(-1, 1, 400)
y_range = np.linspace(-1, 1, 400)

cluster_grid = np.zeros((len(x_range), len(y_range)), dtype=int)

for i, x in enumerate(x_range):
    for j, y in enumerate(y_range):
        if x*x + y*y > 1:
            cluster_grid[j, i] = 8
            continue
        data_point = np.array([x, y]).reshape(1, -1)
        cluster = kmeans_valaro.predict(data_point)
        cluster_grid[j, i] = cluster
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.set_xlabel('valence')
ax.set_ylabel('arousal')
cmap = cm.get_cmap('gist_ncar')
ax.set_box_aspect(1)
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.scatter(valaro_df.values[:, 0], valaro_df.values[:, 1], s=3, c=val_dataset.df['expression'], cmap=cmap, alpha=1, vmin=0, vmax=8,)
ax.contour(x_range, y_range, cluster_grid, colors='black')
fig
```
