```python
from torchaffectnet import AffectNetDataset

train_dataset = AffectNetDataset('../../Affectnet/training.csv',
                                 '../../Affectnet/Manually_Annotated/Manually_Annotated_Images/')
val_dataset = AffectNetDataset('../../Affectnet/validation.csv',
                               '../../Affectnet/Manually_Annotated/Manually_Annotated_Images/')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from tqdm import tqdm

from utils import exclude_id

id2label, _ = exclude_id((8,9,10))

def dataset_plot(df: pd.DataFrame):
    fig = plt.figure()
    ax = fig.add_subplot()
    ax.set_xlabel('valence')
    ax.set_ylabel('arousal')
    cmap = cm.get_cmap('gist_ncar')
    ax.set_box_aspect(1)
    label2point = {}
    for idx, row in tqdm(df.iterrows()):
        mp = ax.scatter(row['valence'], row['arousal'],
                        alpha=1,
                        label=id2label[row['expression']],
                        c=row['expression'],
                        cmap=cmap,
                        vmin=0,
                        vmax=len(id2label),
                        s=3,)
        label2point[id2label[row['expression']]] = mp
    labels, handles = zip(*sorted(label2point.items()))
    legend = ax.legend(handles, labels, loc='upper left', bbox_to_anchor=(1,0.5))
    return fig, legend
```

```python
fig, legend = dataset_plot(val_dataset.df)
fig.savefig(os.path.join('./results/', 'validation_distribution.svg'), bbox_extra_artists=[legend], bbox_inches='tight')
fig.savefig(os.path.join('./results/', 'validation_distribution.png'), bbox_extra_artists=[legend], bbox_inches='tight')
```

```python
fig, legend = dataset_plot(train_dataset.df)
fig.savefig(os.path.join('./results/', 'train_distribution.svg'), bbox_extra_artists=[legend], bbox_inches='tight')
fig.savefig(os.path.join('./results/', 'train_distribution.png'), bbox_extra_artists=[legend], bbox_inches='tight')
```
