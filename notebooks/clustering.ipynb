```python
from itertools import permutations

def generate_combinations(list1, list2):
    if len(list1) != len(list2):
        raise ValueError("Input lists must have the same length")
    
    n = len(list1)
    indices = list(range(n))
    combinations = []
    
    for perm in permutations(indices):
        combination = {list1[i]: list2[perm[i]] for i in range(n)}
        combinations.append(combination)
    
    return combinations

list1 = list(range(8))

combinations = generate_combinations(list1, list1)

print(len(combinations))

# for combination in combinations:
#     print(combination)
```

```python
def calc_kappa(comb, labels_from, labels_to):
    n_match = 0
    for n_samples, (from_cat, to_cat) in enumerate(zip(labels_from, labels_to)):
        if to_cat == comb[from_cat]:
            n_match += 1
    score = n_match / (1 + n_samples)
    return score
```

```python
from tqdm import tqdm

def search_matching(indices, labels_from, labels_to):
    kappa = -1
    for c in tqdm(generate_combinations(indices, indices)):
        score = calc_kappa(c, labels_from, labels_to)
        if score > kappa:
            kappa = score
            matching = c
    return kappa, matching
```

```python
kappa, matching = search_matching(list(range(8)), kmeans_valaro.labels_, kmeans_token.labels_)
```
