```python
from sklearn.cluster import KMeans

num_clusters = 8

kmeans_valaro = KMeans(n_clusters=num_clusters, random_state=0)
kmeans_token = KMeans(n_clusters=num_clusters, random_state=0)
```

```python
import pandas as pd

df = pd.read_csv('./data/CL_expression_after_valaro_z64_bs512token_and_valaro.csv', header=None)
df
```

```python
df_valaro = df.iloc[:, -2:]
df_token = df.iloc[:, :-2]
```

```python
kmeans_valaro.fit(df_valaro)
kmeans_valaro.labels_
```

```python
kmeans_token.fit(df_token)
kmeans_token.labels_
```

```python
from itertools import permutations

def generate_combinations(list1, list2):
    if len(list1) != len(list2):
        raise ValueError("Input lists must have the same length")
    
    n = len(list1)
    indices = list(range(n))
    combinations = []
    
    for perm in permutations(indices):
        combination = {list1[i]: list2[perm[i]] for i in range(n)}
        combinations.append(combination)
    
    return combinations

list1 = list(range(8))

combinations = generate_combinations(list1, list1)

print(len(combinations))

# for combination in combinations:
#     print(combination)
```

```python
from tqdm import tqdm

indices = list(range(8))
kappa = -1
for c in tqdm(generate_combinations(indices, indices)):
    n_match = 0
    for n_samples, (valaro_cat, token_cat) in enumerate(zip(kmeans_valaro.labels_, kmeans_token.labels_)):
        if token_cat == c[valaro_cat]:
            n_match += 1
    score = n_match / (1 + n_samples)
    if score > kappa:
        kappa = score

print('score:', kappa)
```
