```python
import torch
from visualizer import CLS_tokens, plot_tokens_category
from transformers import ViTForImageClassification, ViTFeatureExtractor
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
feature_extractor = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224-in21k')
model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224-in21k').to(device)
```

```python
from torchaffectnet import AffectNetDataset
from torchvision.transforms import Compose, Resize, ToTensor, Normalize

normalize = Normalize(mean=feature_extractor.image_mean, std=feature_extractor.image_std)

transform = Compose([
    Resize(tuple(feature_extractor.size.values())),
    ToTensor(),
    normalize,
])

dataset = AffectNetDataset('../../Affectnet/validation.csv',
                           '../../Affectnet/Manually_Annotated/Manually_Annotated_Images/',
                           transform=transform)
```

```python
from visualizer import hidden_tokens

tokens_list, labels = hidden_tokens(model.to(device), dataset, device)
```
